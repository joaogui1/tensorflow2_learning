```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
def binary_encode(i, num_digits):
    return np.array([i >> d & 1 for d in range(num_digits)])
```

```python
def fizz_buzz_encode(i):
    if   i % 15 == 0: return np.array([0, 0, 0, 1])
    elif i % 5  == 0: return np.array([0, 0, 1, 0])
    elif i % 3  == 0: return np.array([0, 1, 0, 0])
    else:             return np.array([1, 0, 0, 0])
```

```python
NUM_DIGITS = 10
trX = np.array([binary_encode(i, NUM_DIGITS) for i in range(101, 2 ** NUM_DIGITS)])
trY = np.array([fizz_buzz_encode(i)          for i in range(101, 2 ** NUM_DIGITS)])
```

```python
model = keras.Sequential([keras.layers.Dense(100, activation='relu', input_shape=(10,)),
                         keras.layers.Dense(4, activation='softmax')])
```

```python
sgd = keras.optimizers.SGD(lr=0.05)
model.compile(optimizer=sgd,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cb = [keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=1000, restore_best_weights=True)]
```

```python
model.fit(trX, trY, epochs=10000, batch_size=128, validation_split=0.2, callbacks=cb)
```

```python
model.evaluate(trX, trY, verbose=0)[1]
```

```python
def fizz_buzz(i, prediction):
    return [str(i), "fizz", "buzz", "fizzbuzz"][prediction]
```

```python
numbers = np.arange(1, 101)
teX = np.transpose(binary_encode(numbers, NUM_DIGITS))
```

```python
teY = model.predict(teX)
predictions = np.argmax(teY, axis=1)
output = np.vectorize(fizz_buzz)(numbers, predictions)

print(output)
```
